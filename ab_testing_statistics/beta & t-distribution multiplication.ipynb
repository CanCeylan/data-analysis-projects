```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = (12,6)
import random
```

```python
bookers = 2371
visitors = 229985
a = bookers
b = visitors - bookers
```

```python
beta_values = np.random.beta(a,b,10000)

plt.hist(beta_values, bins='auto')
plt.title("Histogram with 'auto' bins")
```

```python
from scipy.stats import beta
import scipy.stats as stats

x = np.linspace(beta.ppf(0.01, a, b),
              beta.ppf(0.99, a, b), 100)
plt.plot(x, beta.pdf(x, a, b),
         'r-', lw=2, alpha=0.6, label='beta pdf')
```

```python
random.choice(beta_values)
```

```python
# An illustration of the t-distribution. Smaller values of t lead to heavier tails (outliers).
x_fit = np.linspace(-6, 6, 1000)
fig, ax = plt.subplots()
t_vals = [1, 2, 100]
sk_vals  = [-2, -1, 1]
for _, sk in zip(t_vals, sk_vals):
  pdf = stats.nct(1, sk, 0, 1).pdf(x_fit)
  ax.plot(x_fit, pdf, linewidth=2)
ax.legend(sk_vals)
```

```python
fig, ax = plt.subplots()
pdf = stats.nct(1, 1, 0, 1).pdf(x_fit)
ax.plot(x_fit, pdf, linewidth=2)
ax.legend('1')
```

```python
random.choice(pdf)
```
