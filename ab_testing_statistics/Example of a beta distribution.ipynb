```python
import numpy as np
from scipy.stats import beta
from matplotlib import pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = (12,6)
```

```python
alpha_values = [0.5, 1.5, 3.0, 0.5]
beta_values = [0.5, 1.5, 3.0, 1.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(0, 1, 1002)[1:-1]
```

```python
# http://www.astroml.org/book_figures/chapter3/fig_beta_distribution.html
fig, ax = plt.subplots()

for a, b, ls in zip(alpha_values, beta_values, linestyles):
    dist = beta(a, b)
    plt.plot(x, dist.pdf(x), ls=ls, label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))
    
plt.xlim(0, 1)
plt.ylim(0, 3)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\alpha,\beta)$')
plt.title('Beta Distribution')

plt.legend(loc=0);
```

```python
# http://varianceexplained.org/statistics/beta_distribution_and_baseball/
a =  81
b = 219
x = np.arange (0.01, 1, 0.01)
y = beta.pdf(x,a,b)
plt.plot(x,y);
```

```python
num_bins = 50
counts, bins = np.histogram(y, bins=num_bins)
bins = bins[:-1] + (bins[1] - bins[0])/2
probs = counts/float(counts.sum())
plt.bar(bins, probs, 1.0/num_bins)
plt.show()
```
